### DATA CLEANING 

```python
#check for null values
pakistan.isnull().sum()
```

```python
#deal with the nulls 
df = pakistan.dropna(subset=['Customer ID', 'Customer Since','sku', 'category_name_1','sales_commission_code'])

mode_A = df['status'].mode()[0]  
df['status'].fillna(mode_A, inplace=True)  
```

```python
df.isnull().sum()
```

The data is now clean for use

```python
# drop the M-Y column since its a repetition of the month and year columns. Also drop the sales_commission_code column since its not essensial for clustering.
df = df.drop(columns=['sales_commission_code','increment_id','FY'], axis=1)
```

### EXPLORATORY DATA ANALYSIS

```python
df['status'].nunique()
```

```python
fig, ax = plt.subplots(figsize=(20, 12))
sns.countplot(data=df, x='status')

plt.xlabel('Status')
plt.ylabel('Count')
plt.title('Count Plot of Status')

# Show the plot
plt.show()
plt.tight_layout()
```

Most of the transactions were completed

```python
df['category_name_1'].nunique()
```

```python
#sort the dataframe by 'price' column in descending order

df = df.sort_values(by='price', ascending=False)

fig, ax = plt.subplots(figsize=(12, 8))

sns.barplot(x = 'category_name_1', y = 'price', data = df, ax = ax).set(title = 'Product Category Vs Average Price')

ax.bar_label(ax.containers[-1], fmt = 'Mean:\n%.2f', label_type = 'edge')

ax.set(xlabel = 'Product Categories')

ax.set(ylabel = 'Price')

ax.set_xticklabels(ax.get_xticklabels(), rotation = 30, horizontalalignment = 'right')

plt.show()
```

The entertainment category had the highest price with a mean of 19295.19. Meaning most customers made there transactions in the entertainment category.

```python
#sort the dataframe by 'price' column in descending order

df = df.sort_values(by='qty_ordered', ascending=False)

fig, ax = plt.subplots(figsize=(12, 8))

sns.barplot(x = 'category_name_1', y = 'qty_ordered', data = df, ax = ax).set(title = 'Product Category Vs Quantity Ordered')

ax.bar_label(ax.containers[-1], fmt = 'Count:\n%.2f', label_type = 'edge')

ax.set(xlabel = 'Product Categories')

ax.set(ylabel = 'Quantity Ordered')

ax.set_xticklabels(ax.get_xticklabels(), rotation = 30, horizontalalignment = 'right')

plt.show()
```

The superstore category had the highest count of transaction.

```python
# Group the data by CustomerID and sum the transaction amounts for each customer
customer_transactions = df.groupby('Customer ID')['price'].sum()

# Find the customer(s) with the highest total transaction amount
max_transaction_customer = customer_transactions.idxmax()
max_transaction_amount = customer_transactions.max()

# Print the result
print(f"Customer {max_transaction_customer} transacted the most with a total of ${max_transaction_amount}")
```

The customer with the most transaction is customer ID 10654, while the highest transaction is $13591207
